```python
import plotly.offline as pyo
import plotly.express as px
import plotly.graph_objs as go
import numpy as np
import pandas as pd
import numpy as np
import warnings 
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
data = pd.read_csv('sentiment_score_final.csv')

data['political_lean_rescaled'].replace('', np.nan, inplace=True)
data.dropna(subset=['political_lean_rescaled'], inplace=True)

data['text'] = '<a href=\"' +  data['Url'] + '\">' + '_</a>'.format("Text")

df = pd.DataFrame({"x":data['pos'].to_list(),
                    "y":data['neg'].to_list(),
                   "hover":data['Title'].to_list(),
                   "compound":data['compound'].to_list(),
                   "engagement":data['Total Engagement'].to_list(),
                   "share":data['Total Shares'].to_list(),
                   "conv":data['Total Conversations'].to_list(),
                   "lean":data['political_lean_rescaled'].to_list(),
                   "topic":data['topic'].to_list(),
                   "source":data['Domain'].to_list(),
                   "text":data['text'].to_list()})
```

```python
df.head()
```

```python
layout = go.Layout(
    title = 'Alexa Top Website Sentiment Score', # Graph title
    xaxis = dict(title = '<---Negative Sentiment-----0-----Positive Sentiment---->'), # x-axis label
    yaxis = dict(title = '<---Political Lean---->'), # y-axis label
)

fig =  go.Figure(layout=layout)

fig.add_trace(
    go.Scatter(x=df["compound"],
               y=df['lean'],
               mode="markers+text",
               # Just pick one of the two
               hovertext=df["hover"],
               text=df["text"],
               textposition="top center",
               textfont_size=4,
              marker=dict(
        size=7,
        color=df["lean"], #set color equal to a variable
        colorscale='balance', # one of plotly colorscales
        showscale=True
   ))
)



fig.show()
```

```python
layout = go.Layout(
    title = 'Alexa Top Website Sentiment Score', # Graph title
    xaxis = dict(title = '<---Positive Score---->'), # x-axis label
    yaxis = dict(title = '<---Negative Score---->'), # y-axis label
)

fig =  go.Figure(layout=layout)

fig.add_trace(
    go.Scatter(x=df["x"],
               y=df["y"],
               mode="markers+text",
               # Just pick one of the two
               hovertext=df["hover"],
               text=df["text"],
               textposition="top center",
               textfont_size=4,
              marker=dict(
        size=7,
        color=df["lean"], #set color equal to a variable
        colorscale='balance', # one of plotly colorscales
        showscale=True
   ))
)



fig.show()
```

```python
layout = go.Layout(
    title = 'Alexa Top Website Popularity vs Sentiment Score', # Graph title
    xaxis = dict(title = '<---Negative Sentiment-----0-----Positive Sentiment---->'), # x-axis label
    yaxis = dict(title = '<---Popularity Score---->'), # y-axis label
)

fig =  go.Figure(layout=layout)

fig.add_trace(
    go.Scatter(x=df["compound"],
               y=df["engagement"],
               mode="markers+text",
               # Just pick one of the two
               hovertext=df["hover"],
               text=df["text"],
               textposition="top center",
               textfont_size=4,
              marker=dict(
        size=7,
        color=df["lean"], #set color equal to a variable
        colorscale='balance', # one of plotly colorscales
        showscale=True
    ))
)
fig.show()
```
